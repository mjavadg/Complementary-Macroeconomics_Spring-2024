## درباره کنترل ورژن

کنترل ورژن سیستمی است که تغییرات اعمال شده بر یک فایل یا مجموعه ای از فایل ها در طول زمان ثبت می کند. استفاده از سیستم کنترل ورژن اجازه می دهد تا ورژن های مختلف فایل ذخیره شود و مورد بررسی واقع شوند و در صورت نیاز تغییرات بازنشانی شوند.

یکی از روش های رایج برای کنترل ورژن فایل ها کپی کردن آن ها در مسیرهای دیگر کامپیوتر است که در شکل زیر مدل آن آمده است. از ایرادات وارد بر این روش این است که مسیر مد نظر ممکن است فراموش است و تغییرات بر فایل اشتباهی ذخیره شود

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread(r'\Screenshot85.png')

plt.imshow(img)
plt.show()
```

روش دیگر استفاده از سیستم کنترل ورژن مرکزی است که برای همکاری با سایرین ایجاد شد. در این روش یک سیستم مرکزی تمام فایل ها و نسخه ها را دارد و در اختیار سایرین می گذارد. در این روش اثر هر فرد بر فایل ها مشخص است و مدیران سیستم کنترل بیشتری در فایل ها و دسترسی سایر افراد دارند. 
از بدی های این سیستم این است که اگر سیستم از دست برود یا سرورهایش خاموش شوند دسترسی تمام افراد به فایل ها از دست می رود.
شکل این سیستم در زیر آمده است. 

```python
img1 = mpimg.imread(r'\Screenshot (87).png')
plt.imshow(img1)
plt.show()
```

روش سوم سیستم کنترل ورژن توزیع شده است که هر کاربر برعکس سیستم مرکزی به کل فایل ها و ورژن های فایل ها دسترسی دارند و می توانند کل فایل ها  و تاریخه شان در سیستم داشته باشند. بنابراین اگر یک سرور یکی از سیستم ها خاموش شوند کاربران همچنان می توانند به فایل ها دسترسی داشته باشند. از سیستم های توزیع شده این چنینی می توان به گیت اشاره کرد که در گیت هاب کاربرانی که از سیستم کنترل ورژن توزیع یافته استفاده می کنند فایل های خود را مدیریت می کنند. مدل این سیستم در شکل زیر آمده.

```python
img2 = mpimg.imread(r'\distributed.png')
plt.imshow(img2)
plt.show()
```

# کار با ریپازیتوری ها

ریپازیتوری ها محل نگهداری فایل ها هستند. می توان یک یا چند نفر به یک ریپازیتوری دسترسی داشته باشند و بر پروژه ای کار کنند یا از فایل های موجود در ریپازیتوری بهره ببرند. 
ابتدا کاربری ریپازیتوری را ایجاد می کند و سپس سایر کاربران فایل های آن را در سیستم خود دریافت می کنند. 
به صورت کلی به دریافت کردن کل ریپازیتوری در یک سیستم کلون کردن و به ارسال فایل از سیستم به ریپازیتوری پوش کردن و به آپدیت کردن فایل های ریپازیتوری در سیستم پول کردن می گویند. 

## کلون کردن از ریپازیتوری

برای اینکه فایل های یک ریپازیتوری در سیستم خود داشته باشیم. ابتدا باید گیت را نصب کنیم که از لینک زیر قابل دریافت است

https://git-scm.com/download/win

ریپازیتوری های کلون شده در مسیری مشخص از سیستم (کامپیوتر) ذخیره می شوند. مسیر از پیش تعیین شده را از طریق اجرای کد زیر در محیط گیت (گیت بش) می توان متوجه شد.

```python
pwd
```

برای تغییر دادن مسیر از پیش تعیین شده از کد زیر استفاده می کنیم.

```python
cd <address>
```

سپس برای کلون کردن یک ریپازیتوری در کامپیوتر خود، ابتدا لینک ریپازیتوری را از صفحه ریپازیتوری کپی می کنیم و سپس از آن برای اجرای کد زیر استفاده می کنیم

```python
git clone <repository link>
```

## پوش به شاخه اصلی ریپازیتوری

هر ریپازیتوری شاخه های مختلفی برای اهداف مختلف می تواند داشته باشد و کاربران متفاوت می توانند به شاخه های متفاوت دسترسی داشته باشند. اما به صورت کلی یک ریپازیتوری یک شاخه اصلی دارد و .در مرحله به چگونگی ارسال (پوش) فایل به شاخه اصلی ریپازیتوری می پردازیم

با استفاده از دستور زیر می توانیم پوشه های مختلف محلی که در آن هستیم را ببینیم

```python
ls
```

باید پوشه ای را انتخاب کنیم که مربوط به ریپازیتوری است و در تنها در این صورت است که فایل هایی که ایجاد می کنیم می توانند به ریپازیتوری پوش شوند. بعد از انتخاب آدرس جدید، فایلی را در آن پوشه می سازیم و سپس دستور زیر را اجرا می کنیم

```python
git status 
```

در نتیجه این دستور اسم فایلی که ساختیم به صورت قرمز نشان داده شده است که یعنی فایل در ریپازیتوری نیست و اصلا آماده ارسال هم نیست.

ممکن است بعد از اجرای این کد با ارور زیر برخورد کنید

```python
$ git status
fatal: not a git repository (or any of the parent directories): .git
```

در این صورت ابتدا دستور زیر را اجرا کنید و مجددا دستور گیت استیتس را اجرا نمایید

```python
git init
```

 برای آماده کردن فایل برای ارسال (استیج کردن) از دستور زیر استفاده می کنیم

```python
git add <file name + format of the file>
```

اکنون با اجرای مجدد دستور گیت استیتس فایل مد نظر به صورت سبز نشان داده می شود که یعنی فایل آماده ارسال (پوش شدن) است.

ممکن است فایل ادد نشود و با ارور زیر مواجه شوید

```python
$ git add test1
fatal: pathspec 'filename' did not match any files
```

در این صورت دستور اول که در زیر آمده است را اجرا کرده و با دستور دوم پوشه ریپازیتوری را انتخاب نمایید

```python
1. ls
2. cd
```

 قبل از پوش شدن عملیات باید به نحوی ثبت شود. که در آن توضیحاتی معمولا آورده می شود که در کد زیر آمده

```python
git commit -m "a name for this act"
```

و در پایان پوش می کنیم. 

```python
git push origin main
```

در دستور بالا، اریجین ریپازیتوری است که به آن متصل هستیم و با  نوشتن اریجین در واقع به ریپازیتوری که آن را کلون کردیم پوش می کنیم. برای تغییر دادن آن می توانیم از دستور زیر استفاده  کنیم.

```python
git remote set-url
```

در راهنمایی ها و ارور ها اریجین ممکن است به صورت زیر نشان داده شود

```python
<remote>
```

مِین نیز اسم شاخه ای است که فایل ها را به آن پوش می کنیم. 

## چگونه با گیت شاخه ای جدید بسازیم؟

برای اینکه بدانیم در کدام شاخه ها را ببینیم:

```python
git branch 
```

برای ایجاد شاخه جدید کد های زیر را به ترتیت اجرا می کنیم

```python
1. git branch "new branch"
2. git push origin "new branch"
```

اگر می خواهید در همان شاخه بمانید دستور زیر را اجرا کنید

```python
git push -u origin "new branch"
```

برای تغییر شاخه ها از کد زیر استفاده می کنیم

```python
git checkout "branch name"
```

## پوش به شاخه های دیگر 

ابتدا شاخه ای که می خواهیم به آن فایل بفرستیم را انتخاب می کنیم سپس مانند مراحل پوش به شاخه اصلی اقدام می کنیم

```python
1. git status
2. git add <file name + format of the file>
3. git commit -m "name of this act"
4. git push <remote> <branch name>
```

## کاربرد پول 

از پول برای آپدیت کردن ریپازیتوری در کامپیوتر استفاده می شود 

```python
1. git pull
2. git pull <remote> <branch>
```

# پایتون

پایتون یک زبان برنام نویسی شیءگرا، تفسیری، سطح بالا، و همه منظوره است، که خیدو فان روسوم آن را طراحی کرد است، و اولین بار در سال ۱۹۹۱ منتشر شد است. فلسفه اصلی طراحی پایتون «خوانایی بالای کد» است و نویس های فاصله خالی در آن معنادار هستند و مکرر استفاده می شوند. ساختار زبانی و دیدگاه شیءگرا در پایتون به گون ای طراحی شد است که به برنامه نویس امکان نوشتن کد منطقی و واضح (بدون ابهام) را برای پروژه های کوچک و بزرگ می دهد.

در پایتون مدل های مختلف برنامه نویسی (از جمله شی گرا و برنامه نویسی دستوری و تابع محور) را پشتیبانی می شود و برای مشخص کردن نوع متغیرها از یک سامانهٔ پویا استفاده می شود.

زبان پایتون ۲ رسماً از سال ۲۰۲۰ دیگر پشتیبانی نمی شود و پایتون ۲٫۷٫۱۸ آخرین نسخهٔ پایتون ۲٫۷ و همچنین پایتون ۲ است و ازاین به بعد دیگر هیچ وصلهٔ امنیتی یا بهبود دیگری در این نسخه داده نخواهد شد.با پایان پشتیبانی از پایتون ۲ فقط پایتون ۳ نسخه های بعدی آن پشتیبانی خواهند شد

پایتون پروژه ای آزاد و متن باز توسعه یافته است و توسط بنیاد نرم افزار پایتون مدیریت می گردد.
