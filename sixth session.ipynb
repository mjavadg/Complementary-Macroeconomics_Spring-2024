```python
data = {
    'USA': {2010: 15049, 2011: 15599, 2012: 16254, 2013: 16843, 2014: 17550, 2015: 18206, 2016: 18695, 2017: 19477, 2018: 20533, 2019: 21381, 2020: 21060, 2021: 23315, 2022: 25462},
    'China': {2010: 6087, 2011: 7551, 2012: 8532, 2013: 9570, 2014: 10475, 2015: 11061, 2016: 11233, 2017: 12310, 2018: 13894, 2019: 14280, 2020: 14687, 2021: 17820, 2022: 17963},
    'France': {2010: 2645, 2011: 2865, 2012: 2683, 2013: 2811, 2014: 2855, 2015: 2439, 2016: 2472, 2017: 2595, 2018: 2790, 2019: 2728, 2020: 2639, 2021: 2957, 2022: 2782},
    'Russia': {2010: 1524, 2011: 2045, 2012: 2208, 2013: 2059, 2014: 1363, 2015: 1276, 2016: 1574, 2017: 1657, 2018: 1693, 2019: 1493, 2020: 1836}
}

plt.figure(figsize =(10, 5))
for country, gdp in data.items():
    years = list(gdp.keys())
    values = list(gdp.values())
    plt.plot(years, values, label=country)

plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('GDP comparison')
plt.legend()
plt.show()
```

# Plotting Functions

<div dir = "rtl">
    ترسیم نمودار توابع مشابه ترسیم نمودارهای قبلی است.
    متغیرها تعریف می‌شوند و سپس از دستور
    ()plt.plot
    برای رسمشان استفاده می‌کنیم.
    
    در مثال زیر تابع نمایی تعریف شده که در آن از کاربر عددی گرفته می‌شود که در 
    x
    ذخیره می‌گردد و سپس متغیر
    v
    که مقدار تابع را ذخیره می‌کند تعریف شده است.
    برای ترسیم کافی است آن‌ها در دستوری که ذکر شد اضافه کنیم.

```python
def expp(x):
    return x**2

x = float(input("Enter a Number"))
v = expp(x)

plt.plot(x, v, marker = "o")
plt.xlabel('Input Value')
plt.ylabel('Squared Value')
plt.title('Exponential Function Plot')
plt.show()
```

<div dir = "rtl">
اما ما به دیدن یک نقطه علاقه نداریم و می‌خواهیم رفتار تابع در نقاط مختلف برسی شود. در این حالت باید از روش‌هایی دیگر برای وارد کردن داده برای متغیرها استفاده کنیم.

```python
def expp(x):
    return x**2

num_inputs = int(input("How many numbers would you like to enter? "))
x_values = []

for i in range(num_inputs):
    x = float(input(f"Enter number {i+1}: "))
    x_values.append(x)

y_values = [expp(x) for x in x_values]

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, marker='o')
plt.xlabel('Input Value')
plt.ylabel('Squared Value')
plt.title('Exponential Function Plot')
plt.grid()
plt.show()
```

<div dir = "rtl">
تفاوت اصلی کد قبل با این کد در این قسمت است

سپس متغیر لیستی تعریف می‌شود که باید مقادیر وارده کاربر در آن ذخیره شود که آن

است. 

در گام بعد حلقه‌ای برای ارسال اعداد وارد شده از کاربر به لیست تعریف شده نوشته می‌شود. 
از تابع

در واقع 
range(num_inputs)
در مثال فوق از 0 تا 5 خواهد بود. (از آن پرینت بگیرید تا بهتر متوجه شوید)

در حلقه، متغیری که قرار است وارد لیست

است.

که توسط کاربر دریافت شد وارد لیست
x_values

در پایان باید متغیری که مقادیر تابع را ذخیره می‌کند تعریف شود که آن نیز لیست است.

```python
def prod_func(labour):
    b = 0.2
    c = 0.9
    d = 0.005
    return b ** labour + c * (labour ** 2) - d * (labour **3)
labour = np.arange(1, 125)
OP = prod_func(labour)

def AP_L(labour):
    return prod_func(labour)/labour

APl = AP_L(labour)

def MP_L(labour):
    b = 0.2
    c = 0.9
    d = 0.005
    return b + 2 * c * labour - 3 * d * (labour ** 2)

MPl = MP_L(labour)

plt.figure(figsize=(10, 5))
plt.plot(labour, OP)

plt.figure(figsize=(10, 5))
plt.plot(labour, APl)
plt.plot(labour, MPl)
plt.show()
```

```python
def cubdog(r, g):
    aa = 0.6
    bb = 0.4
    return (r**aa) * (g**bb)

r = np.linspace(1,10, 50)
g = np.linspace(10,20, 50)

R, G = np.meshgrid(r, g)
value_of_cubdog = cubdog(R, G)

fig = plt.figure()
ax = fig.add_subplot(projection = "3d")
ax.plot_surface(R, G, value_of_cubdog)
plt.title("cobb douglas utility function")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

<div dir ="rtl">
    تابع مطلوبیت کاب داگلاس را با دو کالا را در نمودار فوق مطابق با توضیحاتی که قبلتر آمد رسم کرده‌ایم

```python
def cobb_douglas_utility(x1, x2, a, b):
    return x1 ** a * x2 ** b

a = 0.6
b = 0.4

x1 = np.linspace(1, 10, 100)
x2 = np.linspace(1, 10, 100)
X1, X2 = np.meshgrid(x1, x2)

U = cobb_douglas_utility(X1, X2, a, b)

fig, ax = plt.subplots(figsize=(8, 6))
CS = ax.contour(X1, X2, U, levels=20)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_title(f'Cobb-Douglas Utility Function (a={a}, b={b})')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

def cobb_douglas_utility(x1, x2, a, b):
    return x1 ** a * x2 ** b

a = 0.6
b = 0.4

x1 = np.linspace(1, 10, 50)
x2 = np.linspace(1, 10, 50)
X1, X2 = np.meshgrid(x1, x2)

U = cobb_douglas_utility(X1, X2, a, b)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.contourf3D(X1, X2, U, 50, cmap='viridis')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('Utility')
ax.set_title(f'Cobb-Douglas Utility Function (a={a}, b={b})')
plt.show()
```
